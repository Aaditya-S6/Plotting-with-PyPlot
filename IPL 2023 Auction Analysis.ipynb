# 5. Data Visualization

# Distribution of Type of Player

```python
Type=['Batter','Bowler','All-Rounder','Wicket-Keeper']
Numbers=[7,17,18,7]
pl.figure(1,figsize=(10,8))
pl.bar(Type,Numbers,color=['yellow','red','black','green'])
pl.xlabel('Type of Player')
pl.ylabel('Numbers of player')
pl.title('Distribution of Type of Player')
```

Now, using a Pie chart, We can use a pie chart to visualize actual percentage.

```python
Type=['Batter','Bowler','All-Rounder','Wicket-Keeper']
Numbers=[7,17,18,7]
pl.figure(1,figsize=(10,8))
pl.pie(Numbers,labels=Type,autopct="%5.2f%%")
pl.axis("equal")
pl.title("Distribution of type of player")
```

# Distribution Of Players on the Basis of Team

There are 7 teams and to show this distribution we will use 

```python
import matplotlib.pyplot as pl
team=['chennai super king','delhi capital','gujarat titans','kolkata knight riders','lucknow super giants','mumbai indians','kings 11 punjab']
player=[7,5,7,8,10,8,4]
pl.figure(1,figsize=(15,5))
pl.bar(team,player,color=['orange','yellow','orange','red','red','red','yellow'])
```

Now, Using a Plot

```python
pl.figure(1,figsize=(15,5))
pl.plot(team,player,color='red')

```

### Distribution Of Players based on Sold Price in Auction.

```python
import matplotlib.pyplot as plt
price=[50, 20, 100, 50,20,100,50,20,100,50,20,100,50,200,100,50,20,100,
       50,20,100,50,20,100,50,20,100,50,20,100,50,20,100,50,20, 100,
       50,20,100,50,20,100,50,20,100,50,20,100,50]
plt.figure(1,figsize=(15,5))
plt.hist(price,bins=3,histtype='step',color='grey')
plt.title('DISTRIBUTION OF PLAYERS BASED ON SOLD PRICE')

```
